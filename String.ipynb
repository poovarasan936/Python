```python
a="The world is Beautiful"
print(a[4:9])
print(a[:5])
print(a[7:])
print(a[-15:-9])
print(a[::-1])# Inverted or Reverse String
```

```python
a="The sky is mesmerizing "
print("Uppercase:",a.upper())
print("Lowercase:",a.lower())
print(a.strip())#removes spacing in the start and end of a string
print(a.replace("mesmerizing","beautiful"))#It helps to replace the vaues in the string
print(a.split())#To split the string based on the spaces
b="This is unethical , please be mindful of your actions"
print(b.split(','))#Split the string based on the ','
c='open'
d='world'
print(c+d)#String concatenation
```

```python
a="you are a good Person"
print(a.capitalize())# It changes the first character of a string to uppercase
print(a.casefold())#It changes the first letter of a string to lowercase
```

```python
a="The rabbit is running very fast but the turtle is running very slow"
print(a.count('run'))#To count how many times this word comes in a entire string
print(a.count('run',14,20))#to count wordfrom a specified position
print(a.endswith("slow"))#to check whether the suring ends with the word
print(a.startswith("The"))# To check whether the string starts with the word
```
