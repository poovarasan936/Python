```python
l1=["Dogs","cats"]
l2=["Lions","Elephant","Tiger"]
print(l1)
print(l2)
l1.extend(l2)#The extend() methods adds the elements of list to the end of list l1
print(l1)
l1[2],l1[3]=l1[3],l1[2]#swap two elements in a list
print(l1)
l3={"Panther","Cheeetah"}#when using extend we can extend the list with another list ,set ,dictionary, Tuple
l2.extend(l3)
print(l2)
```

```python
a=["Dinesh","Kishore","Nandha","Sakthivel"]
print(a)
a.pop(2)#pop is used with position it will remove the value at that position
print(a)
b=["one","two"]
print(b)
b.pop()#IF pop used without position,the last value will be removed
print(b)
```

```python
a=["apple","orange","banana","grapes","watermelon"]
print(a)
del a#del is used to delete the entire list
b=["one","two","Three","four"]
print(b)
del b[3]
print(b)
b.clear()#clear is used to clear the entire list but the list will still exist
print(b)
```

```python
c=[1,4,3,6,3,65,3,334,27,4,6]
print(c)
c.sort()
print(c)
var1=["apple","dog","cat","bus"]
var1.sort()
print(var1)
var1.sort(reverse=True)
print(var1)
var2=["Apple","bus","lorry","Banana","airplane"]
print(var2)
var2.sort(key=str.lower)
print(var2)
var2.sort(key=str.upper)
print(var2)
```

```python
a=[23,53,2,4,55,35,43,44]
print(a)
a.reverse()#reverse is used to reverse the order of the list
print(a)
b=a.copy()#copy is used to create a copy of the list a and assign it to b
print(b)
var=list(a)#list is used to create a new list from the existing list a and assign it to var
print(var)
var1=var[:]#var1 is used to create a new list from the existing list var and assign it to var1
print(var1)
```
